```python
# importing modules
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
import seaborn as sns


df = pd.read_csv('data/Iris.csv')
print("Data Shape: ", df.shape)
df.head()
```

Data Analysis

```python
# feature matrix
x_orig = df.iloc[:, 1:-1].values

# data labels
y_orig = df.iloc[:, -1:].values

print("Shape of Feature Matrix: ", x_orig.shape)
print("Shape Label Vector: ", y_orig.shape)
```

Data Visualization

```python
# histogram

"""
Histograms allow you to see the distribution of each feature. You can spot any skewness or unusual values.
"""

df.hist(figsize=(12, 10))
plt.suptitle("Histograms of each feature")
plt.show()



# pair plot

"""
Pair plots show relationships between pairs of features. 
It helps in understanding if there is any clear separation between species based on feature pairs.
"""

sns.pairplot(df, hue='Species', markers=["o", "s", "D"])
plt.suptitle("Pair Plot of Features by Species", y=1.02)
plt.show()



# box plots for each feature by species

"""
Box plots can help to identify any differences in medians and spread (IQR) of features across species. 
It can also spot outliers.
"""

features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

for feature in features:
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f"Box plot of {feature} by Species")
    plt.show()



# violin plots

"""
Violin plots are similar to box plots but also show the distribution of the data.
"""
for feature in features:
    sns.violinplot(x='Species', y=feature, data=df)
    plt.title(f"Violin plot of {feature} by Species")
    plt.show()
```
