# Static Arrays

Arrays of **fixed** size.

An Array must be **contiguous**: *"next or together in sequence. the elements are consecutive, without any interruptions from elements that are not part of the subarray"*

- Reading an array: O(1)
- Writing to an array: O(1)

For static arrays, we don't have control over where the data is stored in memory (RAM). If we want to add a value to a static array, we would ideally need to add that value at the next available address in memory, following the addresses of the existing values in the array. However, when a static array is initialized, we cannot deliberately choose the address of a value. The addresses are automatically assigned by the system, and we cannot modify them directly.

Therefore, we cannot ensure that the address of an added value to a static array will follow the addresses of the values in the array. In other words, this is what makes the array `static`.
